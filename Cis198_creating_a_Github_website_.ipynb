### Github Notes

key terms:
  * **Repository** - a container holding all committed *files* and folders
  * **Commit** - saving a file version in the repository
  * **Git** - this is a version control program (i.e a app for saving many versions of files. Ex. Computer Code)
  * **Github** - an online company providing a place for users to save their repositories
